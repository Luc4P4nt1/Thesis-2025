```python
def compute_isc_optimized(df):
    results = []
    bands = ['Delta', 'Theta', 'Alpha', 'Beta', 'Gamma']

    print("Starting optimized ISC computation...")

    # Iterate through each ID, Content, Vision combination
    for (id_val, content, vision), group in df.groupby(['ID', 'Content', 'Vision']):
        band_isc = {}

        for band in bands:
            # Select relevant columns for this band across channels
            band_columns = [col for col in group.columns if f'Mean PSD' in col and band in col]

            # Skip if NA values exist
            if group[band_columns].isna().any().any():
                band_isc[band] = np.nan
                continue

            signals = group[band_columns].to_numpy()

            # Check if any row is constant or contains NaN
            if np.any(np.isnan(signals)) or np.any(np.all(signals == signals[:, [0]], axis=1)):
                band_isc[band] = np.nan
                continue

            isc_values = []
            for i in range(signals.shape[0]):
                current_signal = signals[i, :]
                other_signals = np.nanmean(np.delete(signals, i, axis=0), axis=0)

                # Check for constant or invalid signals
                if np.all(current_signal == current_signal[0]) or np.all(other_signals == other_signals[0]):
                    isc_values.append(np.nan)
                    continue

                # Compute z-scores
                current_z = (current_signal - np.mean(current_signal)) / np.std(current_signal)
                other_z = (other_signals - np.mean(other_signals)) / np.std(other_signals)

                # Compute normalized cross-covariance
                cross_cov = correlate(current_z, other_z, mode='valid')
                norm_cross_cov = cross_cov / (np.linalg.norm(current_z) * np.linalg.norm(other_z))
                isc_values.append(np.max(norm_cross_cov))

            band_isc[band] = np.nanmean(isc_values)

        # Calculate mean ISC across bands
        band_isc['Mean ISC'] = np.nanmean([band_isc[band] for band in bands])
        results.append({
            'ID': id_val,
            'Content': content,
            'Vision': vision,
            **band_isc
        })

    # Create DataFrame from results
    result_df = pd.DataFrame(results)
    print("Optimized ISC computation completed.")
    return result_df

# Example usage
isc_results_optimized = compute_isc_optimized(df_cleaned)
```

```python
import pandas as pd
import numpy as np
from scipy.signal import correlate
from scipy.stats import zscore

# Carica il dataset (adatta il percorso al tuo file)


# Bande di frequenza
bands = ['Delta', 'Theta', 'Alpha', 'Beta', 'Gamma']
channels = [f'Ch{i}' for i in range(1, 8)]

# Per salvare i risultati
isc_results = []

# Raggruppa per ID, Content, e Vision
grouped = df_cleaned.groupby(['ID', 'Content', 'Vision'])

for (id, content, vision), group in grouped:
    # Ottieni il Frame_Num per sincronizzare i dati
    frame_nums = group['Frame_Num'].values

    for band_idx, band in enumerate(bands):
        isc_band = []

        for channel in channels:
            col_name = f'Mean PSD {channel} {band} (dB)'
            signals = group[col_name].values

            # Z-score il segnale
            signals = zscore(signals, nan_policy='omit')

            # Calcola la ISC
            for idx, signal in enumerate(signals):
                # Media dei segnali degli altri soggetti
                others = np.delete(signals, idx)
                mean_others = np.nanmean(others)

                # Cross-correlazione
                if not np.isnan(mean_others):
                    corr = correlate(signal, mean_others, mode='valid')
                    isc_band.append(np.max(corr))

        # Media su tutti i canali
        isc_results.append({
            'ID': id,
            'Content': content,
            'Vision': vision,
            'Band': band,
            'ISC': np.nanmean(isc_band)
        })

# Crea un nuovo DataFrame con i risultati
isc_df = pd.DataFrame(isc_results)

# Salva i risultati
isc_df.to_csv('isc_results.csv', index=False)

print("Calcolo ISC completato!")
```
